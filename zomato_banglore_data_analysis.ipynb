**Data Cleaning**

```python
#removing the "url","address","phone","Unnamed: 0", "location","rest_type" column
df.drop(["url","address","phone","Unnamed: 0", "location","rest_type"],axis=1,inplace=True)
```

```python
#checking if there are any missing values in name
df["name"].unique()
#captalizing the name
df["name"]=df["name"].str.capitalize()
```

```python
#checking if there are any missing values in online_order
df["online_order"].unique()
```

```python
#checking if there are any book_table missing values
df["book_table"].unique()
```

```python
df["rate"].unique()
```

```python
#checking if there are any missing vlaues
df["rate"].unique()
###there are two missing values -,new,nan
df["rate"].value_counts()
#replacing missing values
df["rate"]=df["rate"].replace('NEW',np.nan)
df["rate"]=df["rate"].replace('-',np.nan)
#striping the unwanted data '/5
df["rate"]=df["rate"].str.strip('/5')
#converting str to float
df["rate"]=df["rate"].astype('float')
#calculating the mean
mean_rate=df["rate"].mean()
#replacing the missing value with mean
df["rate"]=df["rate"].fillna(mean_rate)
df["rate"].unique()
```

```python
#checking if there are any null valus in dish liked column
df["dish_liked"].unique()
#calculating the mode
mode_rest=df["dish_liked"].mode()[0]
#filling the nan with the mode
df["dish_liked"]=df["dish_liked"].fillna(mode_rest)
df["dish_liked"].unique()
df.isnull().sum()
```

```python
#checking if there are any null valus in dish liked column
df["cuisines"].unique()
#calculating the mode
mode_rest=df["cuisines"].mode()[0]
#filling the nan with the mode
df["cuisines"]=df["cuisines"].fillna(mode_rest)
df["cuisines"].unique()
df.isnull().sum()
```

```python
#renaming the column name to cost
d={"approx_cost(for two people)":"cost for two"}
df.rename(columns=d,inplace=True)
#replacing the , with ''
df["cost for two"]=df["cost for two"].str.replace(',',"")
#conerting the string to float
df["cost for two"]=df["cost for two"].astype('float')
#calcuting the mean and replacing the nan values with mean
mean_cost=df["cost for two"].mean().astype(int)
df["cost for two"]=df["cost for two"].fillna(mean_cost)
#changing the data type from float to int
df["cost for two"]=df["cost for two"].astype('int')
df["cost for two"].unique()
```

```python
#renaming the listed_in(type) to restType
d={"listed_in(type)":"restType"}
df.rename(columns=d,inplace=True)
df.columns
#
```

```python
#renaming the listed_in(city) to location
d={"listed_in(city)":"location"}
df.rename(columns=d,inplace=True)
df.columns
#
```

```python
#removing if there are any duplicates present
df.drop_duplicates(inplace=True)
```

```python
#resetting the index of the colums
df.reset_index(drop=True,inplace=True)
df.head()
```
