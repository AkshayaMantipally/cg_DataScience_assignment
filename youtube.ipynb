```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("http://raw.githubusercontent.com/Taj1920/Data_Analysis/refs/heads/main/Youtube_analysis/YOUTUBE%20CHANNELS%20DATASET.csv")
```

```python
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python

mean_rank=df['Ranking'].mean().astype(int)
df['Ranking'].fillna(mean_rank,inplace=True)
df["Ranking"]=df["Ranking"].astype(int)
df["Ranking"].unique()
```

```python

mode_user=df['Username'].mode()[0]
df['Username'].fillna(mode_user,inplace=True)
df["Username"].unique()
```

```python
df["Subscribers"].unique()
df["Subscribers"]=df["Subscribers"].str.strip('M')
mode_sub=df["Subscribers"].mode()[0]
df["Subscribers"].fillna(mode_sub,inplace=True)
df["Subscribers"].unique()
```

```python
df["Uploads"] = df["Uploads"].str.replace(',', '')
df["Uploads"] = df["Uploads"].astype(float)
mean_upload = df['Uploads'].mean()
df['Uploads'].fillna(mean_upload, inplace=True)
df["Uploads"].unique()
```

```python
df["Views"].unique()
df["Views"] = df["Views"].str.replace(',', '')
df["Views"] = df["Views"].astype(float)
mean_upload = df['Views'].mean()
df['Views'].fillna(mean_upload, inplace=True)
df["Views"] = df["Views"].astype(int)
df["Views"].unique()
```

```python
df["Country"].unique()
df["Country"].unique()
df["Country"]=df["Country"].str.strip('M')
mode_sub=df["Country"].mode()[0]
df["Country"].fillna(mode_sub,inplace=True)
df["Country"].unique()
```

```python
df.isnull().sum()
```

```python
df.reset_index(inplace=True)
df.drop("index",axis=1,inplace=True)
```

```python
df.head()
```

**data visualiation**

```python
user_country=df.groupby("Country")["Username"].count().reset_index()
plt.plot(user_country["Country"],user_country["Username"],marker='o')
plt.xlabel('Country')
plt.ylabel('Username')
plt.xticks(rotation=90)
plt.show()
```
